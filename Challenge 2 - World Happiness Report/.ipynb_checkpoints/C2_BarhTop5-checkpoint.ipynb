# Top 5 Countries

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('happiness2019.csv')
df.head()
```

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
df_rename = df.rename(columns = {'variable':'Country or region', 'value':'Country'}, inplace = False)
```

```python
df.info()
```

```python
dff = df.rename(columns = {'Country or region':'Country','GDP per capita':'GDP','Social support':'Social', 'Healthy life expectancy':'Healthy','Freedom to make life choices':'Freedom','Generosity':'Generosity','Perceptions of corruption':'Corrupt'})
```

```python
dff.info()
```

```python
gdp = pd.DataFrame({'Country or region': dff.Country, 'GDP per capita': dff.GDP}) 
gdp_sort = gdp.sort_values('GDP per capita', ascending=True) 
gdp_top10 = gdp_sort.tail(10)
gdp_top10.plot.barh(x='Country or region', y='GDP per capita', figsize= (8,6), title=('Top 10 Countries by GDP per capita'))

plt.show()
```

```python
gdp_sort.plot.barh(x='Country or region', y='GDP per capita', figsize= (10,45)) 
plt.show()
```

```python
social = pd.DataFrame({'Country or region': dff.Country, 'Social support': dff.Social}) 
social_sort = social.sort_values('Social support', ascending=False) 
social_sort.plot.barh(x='Country or region', y='Social support', figsize= (10,45)) 
plt.show()
```

```python
health = pd.DataFrame({'Country or region': dff.Country, 'Healthy life expectancy': dff.Healthy}) 
health_sort = health.sort_values('Healthy life expectancy', ascending=False) 
health_sort.plot.barh(x='Country or region', y='Healthy life expectancy', figsize= (10,45)) 
plt.show()
```

```python
gdp = pd.DataFrame({'Country or region': dff.Country, 'GDP per capita': dff.GDP}) 
gdp_sort = gdp.sort_values('GDP per capita', ascending=True) 
gdp_top5 = gdp_sort.tail(5)
gdp_top5.plot.barh(x='Country or region', y='GDP per capita',  color=['black', 'red', 'green', 'blue', 'cyan'],figsize= (7,5), title=('Top 5 Countries by GDP per capita'))

plt.show()
```

```python
social = pd.DataFrame({'Country or region': dff.Country, 'Social support': dff.Social}) 
social_sort = social.sort_values('Social support', ascending=True)
social_top5 = social_sort.tail(5)
social_top5.plot.barh(x='Country or region', y='Social support',  color=['black', 'red', 'green', 'blue', 'cyan'],figsize= (7,5), title=('Top 5 Countries by Social support'))

plt.show()
```

```python
health = pd.DataFrame({'Country or region': dff.Country, 'Healthy life expectancy': dff.Healthy}) 
health_sort = health.sort_values('Healthy life expectancy', ascending=True)
health_top5 = health_sort.tail(5)
health_top5.plot.barh(x='Country or region', y='Healthy life expectancy', color=['black', 'red', 'green', 'blue', 'cyan'],figsize= (7,5), title=('Top 5 Countries by Healthy life expectancy'))

plt.show()
```

```python
freedom = pd.DataFrame({'Country or region': dff.Country, 'Freedom to make life choices': dff.Freedom}) 
freedom_sort = freedom.sort_values('Freedom to make life choices', ascending=True)
freedom_top5 = freedom_sort.tail(5)
freedom_top5.plot.barh(x='Country or region', y='Freedom to make life choices', color=['black', 'red', 'green', 'blue', 'cyan'], figsize= (7,5), title=('Top 5 Countries by Freedom to make life choices'))

plt.show()
```

```python
generosity = pd.DataFrame({'Country or region': dff.Country, 'Generosity': dff.Generosity}) 
generosity_sort = generosity.sort_values('Generosity', ascending=True)
generosity_top5 = generosity_sort.tail(5)
generosity_top5.plot.barh(x='Country or region', y='Generosity', color=['black', 'red', 'green', 'blue', 'cyan'], figsize= (7,5), title=('Top 5 Countries by Generosity'))

plt.show()
```

```python
corrupt = pd.DataFrame({'Country or region': dff.Country, 'Perceptions of corruption': dff.Corrupt}) 
corrupt_sort = corrupt.sort_values('Perceptions of corruption', ascending=True)
corrupt_top5 = corrupt_sort.tail(5)
corrupt_top5.plot.barh(x='Country or region', y='Perceptions of corruption', color=['black', 'green', 'blue', 'cyan', 'red'], figsize= (7,5), title=('Top 5 Countries by Perceptions of corruption'))

plt.show()
```
