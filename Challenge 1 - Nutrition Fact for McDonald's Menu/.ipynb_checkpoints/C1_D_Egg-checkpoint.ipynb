# Average Calories White Egg vs Whole Egg

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
df = pd.read_csv('menu.csv')
df.head()
```

```python
dff = df.copy()
dff['Item'] = dff['Item'].str.lower()
dff.head()
```

```python
egg_df = dff[dff['Item'].str.contains('egg')]
egg_white = egg_df[egg_df['Item'].str.contains('white')]
egg_whole = egg_df[~egg_df['Item'].str.contains('white')]
```
