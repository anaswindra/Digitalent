```python
import matplotlib.pyplot as plt
crispy = df_data[df_data['Item'].str.contains('Crispy Chicken')]
crispy_cal = pd.DataFrame({'Item': crispy.Item, 'Calories': crispy.Calories})
crispy_sort = crispy_cal.sort_values('Calories', ascending=False)
ax = crispy_sort.plot.barh(x='Item', y='Calories', figsize= (10,7))
```

```python
import matplotlib.pyplot as plt
grilled = df_data[df_data['Item'].str.contains('Grilled Chicken')]
grilled_cal = pd.DataFrame({'Item': grilled.Item, 'Calories': grilled.Calories})
grilled_sort = grilled_cal.sort_values('Calories', ascending=False)
ax = grilled_sort.plot.barh(x='Item', y='Calories', figsize= (10,7))
```

```python
avg_crispy = crispy.Calories.mean(axis=0)
print("Average calories on Crispy Chicken category is ", avg_crispy, "Cal.")
```

```python
# Average calories on Crispy Chicken Category
avg_grilled = grilled.Calories.mean(axis=0)
print("Average calories on Grilled Chicken category is ", avg_grilled, "Cal.")
```

```python
avg_cal = pd.DataFrame({'Categories':['Crispy', 'Grilled'], 'Avg Calories': [avg_crispy, avg_grilled]})
ax = avg_cal.plot.bar(x = 'Categories', y = 'Avg Calories', figsize=(7,5))
ax.set_title("Average Calories on Crispy and Grilled Menu Categories (in Cal)")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
