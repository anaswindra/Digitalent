# Combining Columns

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('E0_.csv')
df.head()
```

```python
#Gabung Nama Team
HT = df['HomeTeam']
AT = df['AwayTeam']
getHT = HT.str[0:3]
getAT = AT.str[0:3]
match = getHT + getAT
match
```

```python
#Combine
FTHG = df['FTHG']
FTAG = df['FTAG']
HTHG = df['HTHG']
HTAG = df['HTAG']
Goal = FTHG + FTAG + HTHG + HTAG

HS = df['HS']
AS = df['AS']
Shoot = HS + AS

HST = df['HST']
AST = df['AST']
Target = HST + AST

HF = df['HF']
AF = df['AF']
Foul = HF + AF

HC = df['HC']
AC = df['AC']
Corner = HC + AC

HY = df['HY']
AY = df['AY']
Yellow = HY + AY

HR = df['HR']
AR = df['AR']
Red = HR + AR
Score = Goal + Shoot + Target + Foul + Corner + Yellow + Red
```

```python
dff = pd.DataFrame({'Div':df.Div, 'Match': match, 'Goal': Goal, 'Shoot': Shoot, 'OnTarget': Target, 'Foul': Foul, 'Corner': Corner, 'Yellow': Yellow, 'Red': Red, 'Score': Score})
```

```python
dff.head()
```

```python
dff.to_csv (r'football_E0.csv', index = False, header=True)

print (dff)
```
