# Top 5 Match Liga

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv('football.csv')
df.head()
```

```python
top = pd.DataFrame({'Match': df.Match, 'Score': df.Score}) 
top_sort = top.sort_values('Score', ascending=True) 
top10 = top_sort.tail(10)
top10.plot.barh(x='Match', y='Score', figsize= (8,6), title=('Top 10 Match All Liga by Score'), color=['blue', 'orange', 'green', 'red', 'purple'])

plt.show()
```
